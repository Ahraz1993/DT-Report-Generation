```python
from reportlab.platypus import (
    SimpleDocTemplate,
    Paragraph,
    Table,
    Spacer
)
from reportlab.lib.styles import getSampleStyleSheet

from reportlab.lib import colors
from reportlab.lib.units import inch
from reportlab.platypus import TableStyle
from reportlab.lib.enums import TA_CENTER
from reportlab.lib.styles import ParagraphStyle
```

```python
styles = getSampleStyleSheet()
```

```python
date_style = ParagraphStyle(
    "Date",
    parent=styles["Normal"],
    alignment=TA_CENTER,
    spaceAfter=20
)
```

```python
title_style = ParagraphStyle(
    "Title",
    parent=styles["Heading1"],
    alignment=TA_CENTER,
    fontSize=20,
    spaceAfter=20
)
```

```python
critical_overload=data_copy.loc[(data_copy["Overloading Status"]=="Critical")&(data_copy["DTRating"]>=100),["Sub Division","Sub Station","DTR Name","DTRating","Overloading Percent","Overloading Status"]]
```

```python
warning_overload=data_copy.loc[(data_copy["Overloading Status"]=="Warning")&(data_copy["DTRating"]>=100),["Sub Division","Sub Station","DTR Name","DTRating","Overloading Percent","Overloading Status"]]
```

```python

warning_overload.iloc[:,-1].name.split(" ")[0]
warning_overload.iloc[:,-1].values[0]
```

```python
def generate_report(data):
     
     report_type=data.iloc[:,-1].name.split(" ")[0]
     report_status=data.iloc[:,-1].values[0]
     doc = SimpleDocTemplate(f"{report_type} {report_status}_Report.pdf")
     elements = []
     elements.append(Paragraph(f" {report_type} {report_status} Report", title_style))
     elements.append(
         Paragraph(
             f"Report Generated : {report_time}",
             date_style
         )
     )
    
    
     for subdivision, df in data.groupby("Sub Division"):
          #Dropping cols not needed
         df=df.drop(columns="Sub Division").rename(columns={f"{report_type} Percent": f"{report_type} %",
         f"{report_type} Status": f"{report_type} Stat","DTRating":"kVA Rating"})
    
         # Sub Division Heading
         heading = Table([[subdivision]], colWidths=[8.2*inch])
    
         heading.setStyle(TableStyle([
             ('BACKGROUND', (0,0), (-1,-1), colors.HexColor('#D9D9D9')),
             ('GRID', (0,0), (-1,-1), 1, colors.black),
             ('FONTNAME', (0,0), (-1,-1), 'Helvetica-Bold'),
             ('FONTSIZE', (0,0), (-1,-1), 12),
             ('LEFTPADDING', (0,0), (-1,-1), 6),
         ]))
    
         elements.append(heading)
    
       
    
         # Table
         table_data = [df.columns.tolist()]
         table_data.extend(df.values.tolist())
         table_data.append([
         "",
         "",
         "",
         "",
         f"Total : {len(df)}"
     ])
    
         table = Table(table_data,colWidths=[
             1.8*inch,   # Sub Station
             2.8*inch,   # DTR Name
             0.8*inch,   # DTRating
             1.2*inch,   # Unbalancing %
             1.1*inch    # Status
         ])
         table.setStyle(TableStyle([
         ('BACKGROUND', (0, 0), (-1, 0), colors.lightgrey),
         ('TEXTCOLOR', (0, 0), (-1, 0), colors.black),
         ('FONTNAME', (0, 0), (-1, 0), 'Helvetica-Bold'),
         ('FONTSIZE', (0, 0), (-1, -1), 8),
         ('GRID', (0, 0), (-1, -1), 0.5, colors.black),
         ('ALIGN', (2, 1), (-1, -1), 'CENTER'),
         ('BOTTOMPADDING', (0, 0), (-1, 0), 8),('SPAN', (0,-1), (3,-1)),            # Pehle 4 columns merge
         ('FONTNAME', (4,-1), (4,-1), 'Helvetica-Bold'),
         ('BACKGROUND', (0,-1), (-1,-1), colors.beige),
         ('ALIGN', (4,-1), (4,-1), 'RIGHT')
     ]))
    
         elements.append(table)
    
       
    
     doc.build(elements)
       
```

```python
dataframe_list=[warning_unbalance,critical_unbalance,critical_overload,warning_overload]
```

```python
for item in dataframe_list:
    if(len(item)):
        generate_report(item)
```
